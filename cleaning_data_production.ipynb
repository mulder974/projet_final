```python
import pandas as pd
```

```python
conso_sheets_list = ['Oil Production - Tonnes',
                     'Gas Production - EJ',
                     'Coal Production - EJ',
                     'Nuclear Generation - TWh',
                     'Hydro Generation - TWh',
                     'Solar Generation - TWh',
                     'Wind Generation -TWh',
                     'Geo Biomass Other - TWh',
                     'Biofuels Production - PJ']

def excel_to_dataframe(sheet:str):
  df = pd.read_excel('data_brut/bp-stats-review-2020-all-data.xlsx',
                    sheet_name=sheet,
                    skiprows=[0,1], skipfooter=8)
  df = df.drop(['2019.1','2008-18','2019.2'],axis=1)
  df = df.transpose()
  new_header = df.iloc[0]
  df = df[1:]
  df.columns = new_header
  df = df.dropna(1,how='all')
  df = df.dropna(0, how='all')
  return df

data_list = []

for sheet in conso_sheets_list:
  clean_df = excel_to_dataframe(sheet)
  data_list.append(clean_df)
```

```python
#Conversion des valeurs en EJ

def Mtonne_to_EJ(x):
    return x * 0.0419

def TWh_to_EJ(x):
    return x * 0.0036

def PJ_to_EJ(x):
    return x * 0.001
    
data_list[0]=Mtonne_to_EJ(data_list[0])

data_list[3] = TWh_to_EJ(data_list[3])
data_list[4] = TWh_to_EJ(data_list[4])
data_list[5] = TWh_to_EJ(data_list[5])
data_list[6] = TWh_to_EJ(data_list[6])
data_list[7] = TWh_to_EJ(data_list[7])
data_list[8] = PJ_to_EJ(data_list[8])



```
